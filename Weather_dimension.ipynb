```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
#longitude and latitude corresponding to the phu
a1 = pd.read_csv("Ottawa_weather.csv")
a2 = pd.read_csv("Toronto_weather.csv")

weather = pd.concat([a1, a2])
```

```python
print(weather.columns)

#limit the date between march and june
start_date = '2020-03-01'
end_date = '2020-06-30'
mask = (weather['Date/Time'] >= start_date) & (weather['Date/Time'] <= end_date)
weather = weather.loc[mask]
#remove duplicate
weather.drop_duplicates(inplace=True)
weather['Climate ID'] = weather['Climate ID'].astype(str)+ "+" +weather['Date/Time'].astype(str)
weather.to_csv("weather.csv",index=False)
weather
```

```python
#Weather dimension
Weather_dimension = pd.DataFrame(columns=['weather_ID', 'max_temp', 'min_temp', 'total_precipitation'])
```

```python
#Mobility Staging
Weather_dimension['weather_ID'] = weather['Climate ID']
Weather_dimension['max_temp'] = weather['Max Temp (°C)']
Weather_dimension['min_temp'] = weather['Min Temp (°C)']
Weather_dimension['total_precipitation'] = weather['Total Precip (mm)']



#Create Surrogate Keys
Weather_dimension.insert(0, "weather_key",  np.arange(len(Weather_dimension)))
#Show Dimension
Weather_dimension.head(10)
```

```python
#Create csv file for Weather dimension
Weather_dimension.to_csv("Weather_dimension.csv",index=False)
```

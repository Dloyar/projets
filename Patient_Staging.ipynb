```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
positive_cases= pd.read_csv("covid_patients.csv")
print(positive_cases.tail(3))
```

```python
print(positive_cases.columns)
#Only get Ottawa and Toronto cases
cities = ["Ottawa", "Toronto"]
positive_cases = positive_cases[positive_cases.Reporting_PHU_City.isin(cities)]


#limit the date between march and june
start_date = '2020-03-01'
end_date = '2020-06-30'
mask = (positive_cases['Case_Reported_Date'] > start_date) & (positive_cases['Case_Reported_Date'] <= end_date)
positive_cases = positive_cases.loc[mask]
positive_cases
```

```python
#Patient dimension
Patient_dimension = pd.DataFrame(columns=['patient_key','patient_ID', 'age_group', 'gender', 'case_acquisitionInfo','outbreak_related'])
```

```python
#patient staging
Patient_dimension['patient_ID'] = positive_cases['Row_ID']
Patient_dimension['age_group'] = positive_cases['Age_Group']
Patient_dimension['gender'] = positive_cases['Client_Gender']
Patient_dimension['case_acquisitionInfo'] = positive_cases['Case_AcquisitionInfo']
Patient_dimension['outbreak_related'] = positive_cases['Outbreak_Related']


#Create Surrogate Keys
key = 100
for index, row in Patient_dimension.iterrows():    
   Patient_dimension.at[index, "patient_key"] = key
   key += 1

#Show Dimension
Patient_dimension
```

```python
#Create csv file for Patient dimension
Patient_dimension.to_csv("Patient_dimension.csv",index=False)
```

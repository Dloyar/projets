```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

Date Dimension Staging

```python
Date_dimension = pd.DataFrame(columns=['date_key','date_ID', 'date', 'day', 'month','year', 'month_name', 'day_of_week_name', 'week_num',
                                       'weekend', 'holiday', 'season', 'day_in_week','day_in_year'])
```

```python
#Date Staging
Date_dimension['date'] = pd.date_range("2020-01-01", "2021-03-30")
Date_dimension["day"] = Date_dimension.date.dt.day
Date_dimension["month"] = Date_dimension.date.dt.month
Date_dimension["year"] = Date_dimension.date.dt.year
Date_dimension["month_name"] = Date_dimension.date.dt.month_name() 
Date_dimension["day_of_week_name"] = Date_dimension.date.dt.day_name()
Date_dimension["week_num"] = Date_dimension.date.dt.isocalendar().week

Date_dimension["day_in_week"] = Date_dimension.date.dt.dayofweek
Date_dimension["day_in_year"] = Date_dimension.date.dt.dayofyear
```

```python
#Holidays of 2020
holidays_in_year_2020 = [1, 48, 101, 139, 183, 251, 286, 360, 363]
holidays_in_year_2021 = [1, 46, 92, 144, 182, 249, 284, 359, 362]

for index, row in Date_dimension.iterrows():
    
    #Staging weekend row
    Date_dimension.at[index, "date_ID"] = index + 1;
    
    #Staging surrogate key (date_key)
    Date_dimension.at[index, "date_key"] = index + 9000;
    
    if row["day_in_week"] > 4:
        Date_dimension.at[index, "weekend"] = 1
    else:
        Date_dimension.at[index, "weekend"] = 0
    
    #Staging is_holiday row
    is_holiday = False;
    for holiday in holidays_in_year_2020:
        if holiday == row["day_in_year"]:
            is_holiday = True
            break
    
    if is_holiday:
        Date_dimension.at[index, "holiday"] = 1
    else:
        Date_dimension.at[index, "holiday"] = 0

    #Staging season row
    if row["day_in_year"] in range(80, 172):
        Date_dimension.at[index, "season"] = "Spring"
    elif row["day_in_year"] in range(172, 265):
        Date_dimension.at[index, "season"] = "Summer"
    elif row["day_in_year"] in range(265, 355):
        Date_dimension.at[index, "season"] = "Fall"
    else:
        Date_dimension.at[index, "season"] = "Winter"
```
